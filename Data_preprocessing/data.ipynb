```python
import os
import pandas as pd
from PIL import Image
import numpy as np
```

### Обработка и разметка изображений для обучения

```python
plants_list = ['Dracaena', 'Hovea', 'Orchid', 'Sansevieria', 'Spathiphyllum',
               'Zamiokulkas', 'Aglaonema', 'Kalanchoe']
```

```python
plant_data = pd.read_csv('data/new_plants_data.csv')

plants_list = ['Cactus', 'Ficus_Elastica']

n = 8

save_path = 'data/Plants_data_for_ML/train/'

for i in plants_list:
    
    path = 'data/Raw_data/' + i + '/train'
    
    for j in os.listdir(path):
        
        if str(j)[-3:] == 'jpg' or str(j)[-3:] == 'png' or str(j)[-4:] == 'jpeg': 
            
            img_list = []
            img_index = [n] * 13
            
            with Image.open(path + '/' + j) as img:
                img.load()
                
            img_name = j
            img_list.append(img_name)
            img.save(save_path + img_name)
            
            gray_img = img.convert("L")
            img_name = 'gray_' + j
            img_list.append(img_name)
            gray_img.save(save_path + img_name)
            
            conv_img = img.transpose(Image.FLIP_TOP_BOTTOM)
            img_name = 'conv_' + j
            img_list.append(img_name)
            conv_img.save(save_path + img_name)
            
            gray_conv_img = conv_img.convert("L")
            img_name = 'gray_conv_' + j
            img_list.append(img_name)
            gray_conv_img.save(save_path + img_name)
            
            img_rot90 = img.copy()
            crop = img.width * 0.7
            sub_image = img.crop(box=(100,100,crop,crop)).rotate(90)
            img_rot90.paste(sub_image, box=(100,100))
            img_name = 'rot90_' + j
            img_list.append(img_name)
            img_rot90.save(save_path + img_name)
            
            gray_rot90_img = img_rot90.convert("L")
            img_name = 'gray_rot90_' + j
            img_list.append(img_name)
            gray_rot90_img.save(save_path + img_name)
            
            img_rot180 = img.copy()
            crop = img.width * 0.8
            sub_image = img.crop(box=(50,50,crop,crop)).rotate(180)
            img_rot180.paste(sub_image, box=(50,50))
            img_name = 'rot180_' + j
            img_list.append(img_name)
            img_rot180.save(save_path + img_name)
            
            gray_rot180_img = img_rot180.convert("L")
            img_name = 'gray_rot180_' + j
            img_list.append(img_name)
            gray_rot180_img.save(save_path + img_name)
            
            img_width = img.resize((round(img.width*1.5), img.height))
            img_name = 'width_' + j
            img_list.append(img_name)
            img_width.save(save_path + img_name)
            
            width_gray_img = img_width.convert("L")
            img_name = 'width_gray_' + j
            img_list.append(img_name)
            width_gray_img.save(save_path + img_name)
            
            img_wid_rot90 = img_width.copy()
            crop = img_width.width * 0.7
            sub_image = img_width.crop(box=(100,100,crop,crop)).rotate(90)
            img_wid_rot90.paste(sub_image, box=(100,100))
            img_name = 'wid_rot90_' + j
            img_list.append(img_name)
            img_wid_rot90.save(save_path + img_name)
            
            img_height = img.resize((img.width, round(img.height*1.2)))
            img_name = 'height_' + j
            img_list.append(img_name)
            img_height.save(save_path + img_name)
            
            height_gray_img = img_height.convert("L")
            img_name = 'height_gray_' + j
            img_list.append(img_name)
            height_gray_img.save(save_path + img_name)
            
            interm_df = pd.DataFrame({'img_name': img_list, 'id': img_index})
            plant_data = pd.concat([plant_data, interm_df], ignore_index = True)
            
    n += 1

display(plant_data.tail())
```

```python
# Записать новый файл
plant_data.to_csv('data/Plants_data_for_ML/Plants_data_train.csv', index=False)
```

```python
# Дозаписать в существующий файл
train_plant_data = pd.read_csv('data/Plants_data_for_ML/Plants_data_train.csv')

train_plant_data = pd.concat([train_plant_data, plant_data], ignore_index = True)

train_plant_data.to_csv('data/Plants_data_for_ML/Plants_data_train.csv', index=False)
```
